# Linear Regression

```python
import pandas as pd
```

```python
df=pd.read_csv('Boston_housing.csv',header=None, sep='\\s+', names= ['CRIM', 'ZN', 'INDUS', 'CHAS', 'NOX', 'RM', 'AGE', 'DIS', 'RAD', 'TAX', 'PTRATIO', 'B', 'LSTAT', 'MEDV'])
```

```python
df.head()
```

```python
df.info()
```

```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.figure(figsize=(12, 8))
sns.heatmap(df.corr(),annot=True,fmt='.2f')
```

```python
X = df.drop('MEDV', axis=1)
y = df['MEDV']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression().fit(X_train, y_train)
```

```python
reg
```

```python
pred=reg.predict(X_test)
```

```python
from sklearn.metrics import r2_score
r2_score(y_test, pred)
```

```python
import numpy as np

# Define new data point
new_data = np.array([[0.1, 0.0, 7.0, 0, 0.4, 6.5, 60.0, 5.0, 3, 300, 18.0, 396.0, 5.0]])

# Predict using the trained model
predictions = reg.predict(new_data)

print("Predicted MEDV:", predictions[0])
```
