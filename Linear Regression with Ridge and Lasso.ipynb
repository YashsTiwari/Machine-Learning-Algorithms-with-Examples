## Linear regression

```python
#house Pricing dataset
```

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline 
```

```python
dataset=pd.read_csv('Boston_housing.csv',header=None, sep='\\s+', names= ['CRIM', 'ZN', 'INDUS', 'CHAS', 'NOX', 'RM', 'AGE', 'DIS', 'RAD', 'TAX', 'PTRATIO', 'B', 'LSTAT', 'MEDV'])
dataset = dataset.rename(columns={'MEDV': 'Price'})
```

```python
dataset.head()
```

```python
## Dividing the dataset into independent and dependent features
X=dataset.iloc[:,:-1] #dependent feature
y=dataset.iloc[:,-1] #dependent feature
```

```python
X.head(3)
```

```python
y.head(3)
```

```python
## Linear Regression
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import cross_val_score
lin_reg=LinearRegression()
# better to do train test split and give Xtrain and ytrain
mse=cross_val_score(lin_reg,X,y,scoring='neg_mean_squared_error',cv=5)
mean_mse=np.mean(mse)
print(mean_mse)
```

```python
lin_reg.predict(# give values)
```

## Ridge Regression

```python
from sklearn.linear_model import Ridge
from sklearn.model_selection import GridSearchCV
ridge=Ridge()

params={'alpha':[1e-15,1e-10,1e-8,1e-3,1e-2,1,5,10,20]}
ridge_regressor=GridSearchCV(ridge,params,scoring='neg_mean_squared_error',cv=5)
ridge_regressor.fit(X,y) #give Xtrain, ytrain
```

```python
print(ridge_regressor.best_params_)
print(ridge_regressor.best_score_)
```

Poor performance than linear regression

## Lasso Regression

```python
from sklearn.linear_model import Lasso
from sklearn.model_selection import GridSearchCV
lasso=Lasso()

params={'alpha':[1e-15,1e-10,1e-8,1e-3,1e-2,1,5,10,20]}

lasso_regressor=GridSearchCV(lasso,params,scoring='neg_mean_squared_error',cv=5)
lasso_regressor.fit(X,y) #give Xtrain, ytrain
```

```python
print(lasso_regressor.best_params_)
print(lasso_regressor.best_score_)
```

## Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
from sklearn.datasets import load_breast_cancer
```

```python
df=load_breast_cancer()
# Independent features
X=pd.DataFrame(df['data'],columns=df['feature_names'])
X.head(3)
```

```python
# Dependent feature
y=pd.DataFrame(df['target'],columns=['Target'])
y.head(3)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
X, y, test_size=0.33, random_state=42)
```

```python
params=[{'C':[1,5,10]},{'max_iter':[100,150]}]
```

```python
model1=LogisticRegression(C=100,max_iter=100)
```

```python
model=GridSearchCV(model1,param_grid=params,scoring='f1',cv=5)
```

```python
model.fit(X_train,y_train)
```

```python
model.best_params_
```

```python
model.best_score_
```

```python
y_pred=model.predict(X_test)
y_pred
```

```python
from sklearn.metrics import confusion_matrix,classification_report,accuracy_score
confusion_matrix(y_pred,y_test)
```

```python
accuracy_score(y_test,y_pred)
```

```python
print(classification_report(y_test,y_pred))
```
